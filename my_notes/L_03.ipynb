Quando stimiamo dai dati stiamo facendo "sample statistics" mentre se deriviamo statistica da una popolazione stiamo facendo "population statistics" e in quel caso conosciamo la pdf. 
$\mu$ è la population mean, il valore di aspettazione di x per h(x). Non conoscendo h(x) definiamo $\bar{x}$ "sample mean" ovvero l'estimatore di $\mu$

Invece di conoscere la "population variance $\sigma^2$" ci serve la sample variance $s^2$ dove ci serve un fattore correttivo di $\frac{1}{N-1}$ invece di 1/N chiamato "Bessel Correction". In effetti se N è molto grande questo -1 è trascurabile.


NB: "s" is not the error on the mean, l'incertezza di $\bar{x}$ è lo "standard error of the mean" che è $\sigma_{\bar{x}} = \frac{s}{\sqrt{N}}$. La deviazione standard è la larghezza estimata della distribuzione.
